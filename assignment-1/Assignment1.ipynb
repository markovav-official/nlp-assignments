## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from collections import Counter
import re


class BPETokenizer:
    """
    Byte-Pair Encoding (BPE) tokenizer
    """

    def __init__(self, vocab_size: int = 100) -> None:
        self.vocab_size = vocab_size
        self.vocab = {}
        self.merges = {}

    def train(self, texts: list[str]) -> None:
        """
        Train the tokenizer on a list of texts
        """
        # Initialize the vocabulary with the individual characters in the text (and the end-of-word symbol </w>)
        self.vocab = Counter()
        for text in texts:
            for word in text.lower().split():
                self.vocab[" ".join(list(word)) + " </w>"] += 1

        # Merge the most frequent pairs of characters (num_merges times)
        while self._get_vocab_size() < self.vocab_size:
            # Get the most frequent pairs of characters
            pairs = self._get_stats(self.vocab)
            if not pairs:
                break

            # Get the most frequent pair
            best = max(pairs, key=pairs.get)

            # Merge the most frequent pair
            self.vocab = self._merge_vocab(best, self.vocab)

            # Save the merge
            self.merges[best] = pairs[best]

    def tokenize(self, text: str) -> list[str]:
        """
        Tokenize a text using the trained tokenizer
        """
        # Add the end-of-word symbol to each word and split the words into characters
        words = [" ".join(list(word)) + " </w>" for word in text.split()]
        for i in range(len(words)):
            word = words[i]

            # Replace the most frequent pairs of characters with the merged version
            for merge in self.merges:
                word = re.sub(re.escape(" ".join(merge)), "".join(merge), word)

            words[i] = word

        tokens = []
        for word in words:
            tokens.extend(word.rstrip('</w>').strip().split())
        return tokens

    def _get_vocab_size(self):
        """
        Get the size of the vocabulary
        """
        tokens = set()
        for word in self.vocab:
            tokens.update(word.split())
        return len(tokens)

    def _get_stats(self, vocab):
        """
        Get the most frequent pairs of characters in the vocabulary
        """
        pairs = Counter()
        for word, freq in vocab.items():
            symbols = word.split()
            for i in range(len(symbols) - 1):
                pair = (symbols[i], symbols[i + 1])
                pairs[pair] += freq
        return pairs

    def _merge_vocab(self, pair, v_in):
        """
        Merge the most frequent pair of characters in the vocabulary
        """
        v_out = {}
        bigram = re.escape(" ".join(pair))
        # Create a regular expression to match the most frequent pair of characters
        p = re.compile(r"(?<!\S)" + bigram + r"(?!\S)")
        for word in v_in:
            # Replace the most frequent pair of characters with the merged version
            w_out = p.sub("".join(pair), word)
            v_out[w_out] = v_in[word]
        return v_out
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
files = [f"file{i + 1}" for i in range(5)]

texts = []
for file in files:
    with open(f"data/{file}", "r") as f:
        texts.append(f.read())
```

```python
vocab_sizes = [0, 250, 500, 750, 1000]

average_lengths = []
for vocab_size in vocab_sizes:
    tokenizer = BPETokenizer(vocab_size)
    tokenizer.train(texts)
    if vocab_size == 0:
        vocab_sizes[0] = tokenizer._get_vocab_size()

    lengths = []
    for text in texts:
        for text in text.split("\n"):
            tokens = tokenizer.tokenize(text)
            lengths.append(len(tokens))
    average_lengths.append(sum(lengths) / len(lengths))

print(average_lengths)
```

```python
from matplotlib import pyplot as plt


plt.plot(vocab_sizes, average_lengths)
plt.xlabel("Vocab size")
plt.ylabel("Average length")
plt.title("Average length of tokenized tweet by vocab size")
plt.grid()
plt.show()
```

**Tell what `vocab_size` is preferrable and why.**

**Answer:** \
Choosing the right `vocab_size` for tokenizing tweets is about finding a balance. A `vocab_size` between 500 to 750 is often best. This size is big enough to cover the variety of language in tweets, including slang and emojis, without being too large or too small. A smaller `vocab_size` might not capture all the details well, leading to longer tokenized tweets. On the other hand, a very large `vocab_size` could make the model too complex and less efficient. So, a mid-range `vocab_size` offers a good compromise by being efficient and expressive enough for most tweet-related tasks.
